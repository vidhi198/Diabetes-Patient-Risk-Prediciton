## Title : Diagnosis to Predict Whether a Patient is Likely to Have Diabetes

### Business Problem: Predicting Diabetes Risk

Healthcare providers face the challenge of early diabetes detection. Diagnosing at-risk patients using expert evaluation is time-consuming and not scalable. Can we use predictive analytics to automate this process? Business Problem: How can we replicate expert diagnosis to predict whether a patient is likely to have diabetes, based on measurable health indicators?

### Dependent and Independent Variables

Dependent Variable (Categorical):

- Outcome:

    - 1 = Patient has diabetes

    - 0 = Patient does not have diabetes


Independent Variables (Quantitative):

- Pregnancies: Number of times pregnant

- Glucose: Plasma glucose concentration

- BloodPressure: Diastolic blood pressure (mm Hg)

- SkinThickness: Triceps skinfold thickness (mm)

- Insulin: 2-Hour serum insulin (mu U/ml)

- BMI: Body mass index (weight in kg/(height in m)^2)

- DiabetesPedigreeFunction: Diabetes pedigree function (family history influence)

- Age: Age in years

#### Correlation Matrix and Heatmap

- Purpose and Interpretation of the Correlation Matrix

The **correlation matrix** is a valuable tool in feature selection and exploratory data analysis, especially in the context of predictive modeling. It helps identify how strongly each feature is related to the target variable—in this case, whether a person has diabetes (Outcome)—as well as how features relate to one another. This is important for two reasons: first, features that show a strong correlation with the outcome are typically more useful for building predictive models; second, detecting features that are highly correlated with each other can help avoid redundancy and multicollinearity, which can distort model performance.


```python
# Compute correlation matrix (only numeric columns)
corr = df.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

In the correlation matrix we generated, the feature most strongly correlated with the Outcome is **Glucose**, with a correlation coefficient of approximately **0.47**. This indicates a moderate positive relationship, meaning higher glucose levels are associated with a greater likelihood of having diabetes. This aligns well with medical understanding and suggests that glucose is a critical predictive feature.

**BMI (Body Mass Index)** shows a correlation of about **0.29** with the outcome, indicating that people with higher BMI are somewhat more likely to develop diabetes. Though not as strong as glucose, this is still a meaningful signal and supports the inclusion of BMI in the model. Similarly, **Age** has a correlation of **0.24** with diabetes, suggesting that older individuals are at a higher risk. This is consistent with health data that shows age as a risk factor.

**Pregnancies** has a slightly lower correlation of around **0.22**, but this still indicates that women who have had more pregnancies may be at a somewhat higher risk of diabetes, likely due to factors like gestational diabetes or changes in metabolism over time. The **Diabetes Pedigree Function**, which is a proxy for genetic risk, has a correlation of **0.17**, which is weaker but still relevant, especially considering that genetic predisposition may interact with other features in complex ways.

Some features like **Insulin**, **BloodPressure**, and **SkinThickness** have very weak correlations with the outcome (ranging from about **0.07 to 0.13**). While these may not be strong predictors on their own, they could still contribute value in combination with other variables in multivariate or non-linear models.

Looking at correlations between input features, we can also identify potential redundancy. For instance, **SkinThickness** and **Insulin** are correlated at **0.44**, and **SkinThickness** and **BMI** are correlated at **0.39**. These features all relate to body fat and metabolic health and may carry overlapping information. Additionally, **Pregnancies** and **Age** show a fairly high correlation of **0.54**, which makes intuitive sense as older individuals are more likely to have had more pregnancies. These insights suggest that while all features may contribute to model accuracy, some carry similar information and may be candidates for transformation, dimensionality reduction, or selection based on model performance.

In summary, the correlation matrix confirms that **Glucose**, **BMI**, and **Age** are the most valuable predictors of diabetes in this dataset, while other features may support the model through interactions or non-linear effects. Understanding these relationships supports more informed, efficient, and accurate predictive modeling.

### Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('diabetes.csv')  # Make sure the file is in the same directory or update the path

# Create age bins and labels
bins = [20, 30, 40, 50, 60, 70, 80]
labels = ['20-29', '30-39', '40-49', '50-59', '60-69', '70-79']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Group by AgeGroup and Outcome, then count
age_outcome_distribution = df.groupby(['AgeGroup', 'Outcome']).size().unstack(fill_value=0)

# Plot pie charts for each age group
fig, axes = plt.subplots(2, 3, figsize=(18, 10))  # 6 groups -> 2 rows x 3 columns
colors = ['#ff9999','#66b3ff']  # Colors for outcome 0 and 1
explode = (0.05, 0.05)  # Slightly explode both slices for emphasis

# Generate individual pie charts
for ax, age_group in zip(axes.flatten(), age_outcome_distribution.index):
    sizes = age_outcome_distribution.loc[age_group]
    ax.pie(sizes,
           labels=['Non-Diabetic', 'Diabetic'],
           autopct='%1.1f%%',
           startangle=140,
           colors=colors,
           explode=explode,
           shadow=True)
    ax.set_title(f'Age Group {age_group}')

# Super title and layout adjustments
plt.suptitle('Diabetes Outcome Distribution by Age Group', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

Interpretation: These pie charts reveal the proportion of diabetic vs. non-diabetic patients across segmented age groups. A clear upward trend in diabetes prevalence is observed starting from the 30–39 age group, intensifying through 50–59. This age-linked risk escalation indicates that preventive screenings and targeted interventions should focus on individuals aged 30 and above, with the 50+ demographic being of highest priority for clinical outreach or health programs.

```python
# Set the visual style
sns.set(style="whitegrid")

# Create a figure with two subplots: one for Pregnancies, one for Age
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Bar plot: Average Outcome per number of Pregnancies
sns.barplot(x='Pregnancies', y='Outcome', data=df, ax=axes[0],
            ci=None, palette='coolwarm')
axes[0].set_title('Average Diabetes Outcome by Number of Pregnancies')
axes[0].set_ylabel('Average Outcome (1 = Diabetic)')
axes[0].set_xlabel('Number of Pregnancies')

# Bar plot: Average Outcome per Age
sns.barplot(x='Age', y='Outcome', data=df, ax=axes[1],
            ci=None, palette='coolwarm')
axes[1].set_title('Average Diabetes Outcome by Age')
axes[1].set_ylabel('Average Outcome (1 = Diabetic)')
axes[1].set_xlabel('Age')
axes[1].tick_params(axis='x', rotation=90)

# Layout adjustment
plt.tight_layout()
plt.show()
```

Interpretation: This chart plots the average diabetes outcome (0 = non-diabetic, 1 = diabetic) by pregnancy count. A strong correlation is observed: women with 6+ pregnancies show significantly higher average diabetes outcomes, indicating a risk threshold. From a policy or programmatic standpoint, this suggests a need for proactive monitoring of maternal health, particularly for women with high parity, as part of prenatal and postnatal care initiatives.
